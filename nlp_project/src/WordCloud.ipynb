```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
#Reading the txt file as a dataframe
df = pd.read_csv('/Users/SaberKlai/Documents/Projekt_SentAn/NLP_Project_progressing/src/preprocessed_tweets.txt', delimiter=" \t", engine='python')
#Adding a column with the title "Tweets"
df.columns = ['Tweets']
df
```

```python
comment_words = ''
stopwords = set(STOPWORDS)

for val in df['Tweets']:
    #Typecasting
    val = str(val)
    #Tokenization
    tokens = val.split()
    #Lowercasing
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()

    comment_words += " ".join(tokens)+" "

wordcloud = WordCloud(width = 800, height = 800,
                background_color ='gray',
                stopwords = stopwords,
                min_font_size = 12).generate(comment_words)  
```

```python
#Plotting the word cloud
plt.figure(figsize = (7, 7), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
